# Проект: телекоммуникации

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Задача — обучить на этих данных модель для прогноза оттока клиентов.    

**Описание услуг**    
*Оператор предоставляет два основных типа услуг:*    
- `Стационарную телефонную связь.` Телефон можно подключить к нескольким линиям одновременно.    
- `Интернет.` Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).
     
*Также доступны такие услуги:*    
- `Интернет-безопасность:` антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);    
- `Выделенная линия технической поддержки (TechSupport)`;    
- `Облачное хранилище файлов для резервного копирования данных (OnlineBackup)`;    
- `Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies)`.    
     
Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.    
     
**Описание данных**    
*Данные состоят из нескольких файлов, полученных из разных источников:*    
- `contract_new.csv` — информация о договоре;    
- `personal_new.csv` — персональные данные клиента;    
- `internet_new.csv` — информация об интернет-услугах;

Пропуски необходимо как-то заполнить. Пропуски в данных столбцах означают, что у абонента не подключены данные услуги. Но мы не можем заполнить их просто как "No", так как нужно различать пользователей, у которых не подключена какая-то услуга, и у которых  в приинципе не подключён интернет.

```python
contract_new_merge.isna().sum()
```

Заменил все пропуски на Nan

```python
(contract_new_merge == 'Nan').sum()
```

```python
(contract_new_merge == 'Nan').sum() / contract_new_merge.count()
```

```python
contract_new_merge['leave'] = contract_new_merge['EndDate']

contract_new_merge.loc[contract_new_merge['leave'] != 'No', 'leave'] = 1
contract_new_merge.loc[contract_new_merge['leave'] == 'No', 'leave'] = 0
```

Создал целевой признак `leave`, который говорит ушёл клиент или нет

```python
contract_new_merge['leave'].value_counts().plot(kind='pie', autopct='%1.1f%%');
```

```python
contract_new_merge.info()
```

```python
contract_new_merge['contract_duration'] = ''
```

```python
for i in range(contract_new_merge.shape[0]):
    if contract_new_merge['EndDate'][i] != 'No':
        contract_new_merge['contract_duration'][i] = ((dt.strptime(contract_new_merge['EndDate'][i], "%Y-%m-%d"))\
        - contract_new_merge['BeginDate'][i]).days
    else:
        contract_new_merge['contract_duration'][i] = ((dt.strptime("2020-02-01", "%Y-%m-%d"))\
        - contract_new_merge['BeginDate'][i]).days
```

```python
contract_new_merge['contract_duration']
```

Создал новый признак `contract_duration` - длительность контракта в днях.

## Шаг 5. Подготовка данных

```python
contract_new_merge['leave'] = contract_new_merge['leave'].astype(int)
```

```python
RANDOM_STATE=180324
TEST_SIZE=0.25

features = contract_new_merge.drop(['leave'], axis=1)
target = contract_new_merge['leave']

X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE
)

X_train_cb = X_train.copy()
X_test_cb = X_test.copy()
```

Разделил данные на признаки и целевой признак. А после на обучающую и тестовую выборку.
Создал копию X_train и X_test для модели CatBoostClassifier, так как для неё не нужно кодировать категориальные признаки

```python
ohe_features = X_train.select_dtypes(include='object').columns.to_list()
print(ohe_features)

num_features = X_train.select_dtypes(exclude='object').columns.to_list()
num_features.remove('SeniorCitizen')
num_features
```

Разделил признаки на категориальные и численные

```python
encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
encoder_ohe.fit(X_train[ohe_features])

X_train[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(X_train[ohe_features])

X_train = X_train.drop(ohe_features, axis=1)

scaler = StandardScaler()

X_train[num_features] = scaler.fit_transform(X_train[num_features])

X_train.head()
```

```python
X_test[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(X_test[ohe_features])

X_test = X_test.drop(ohe_features, axis=1)

X_test[num_features] = scaler.transform(X_test[num_features])

X_test.head()
```

Закодировал категориальные признаки и масштабировал количественные

## Шаг 6. Обучение моделей машинного обучения

### LogisticRegression

```python
%%time

logistic_regression = LogisticRegression(random_state=RANDOM_STATE)

logistic_regression.fit(X_train, y_train)
```

Обучил логистическую регрессию на обучающей выборке

```python
cv_scores = cross_val_score(logistic_regression, X_train, y_train, scoring='roc_auc', cv=3)

print('Best model parameters:', logistic_regression.get_params())
print(f'Cross-validation roc_auc score: {cv_scores.mean()}')
```

Посчитал ROC-AUC для LogisticRegression на кросс-валидации

### CatBoost

```python
%%time

catboost = CatBoostClassifier(cat_features=ohe_features, eval_metric='AUC', random_seed=RANDOM_STATE)

grid = {'n_estimators': [100, 1000, 1100],
        'max_depth': [4, 6],
        'learning_rate': [0.0001, 0.03, 0.1]}

grid_search_result = catboost.grid_search(grid,
                                          X=X_train_cb,
                                          y=y_train,
                                          cv=3,
                                          plot=True)
```

```python
grid_search_result["params"]
```

```python
%%time
final_cat = CatBoostClassifier(cat_features=ohe_features,
                              n_estimators=grid_search_result["params"]["iterations"],
                              learning_rate=grid_search_result["params"]['learning_rate'],
                              max_depth=grid_search_result["params"]['depth'],
                              eval_metric='AUC',
                              verbose=100,
                              random_seed=RANDOM_STATE)
final_cat.fit(X_train_cb, y_train)
```

```python
cv_scores = cross_val_score(final_cat, X_train_cb, y_train, scoring='roc_auc', cv=3)

print('Best model parameters:', final_cat.get_params())
print(f'Cross-validation roc_auc score: {cv_scores.mean()}')
```

На кроссвалидации лучший результат показал CatBoostClassifier
